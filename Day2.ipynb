```python
choice = input("Enter Transaction (withdraw/deposit): ")

if choice == "withdraw":
    amount = int(input("Enter the Amount: "))
    
    if amount <= 10000:
        pin = int(input("Enter the PIN: "))
        
        if pin == 2580:
            print("Withdrawal Successful")
        else:
            print("Incorrect PIN")
    else:
        print("Withdrawal Amount should not be greater than 10K")

elif choice == "deposit":
    amount = int(input("Enter the Amount to Deposit: "))
    
    if amount >= 0:
        print("Deposit Successful")
    else:
        print("Invalid Amount")

else:
    print("Invalid Transaction")
```
