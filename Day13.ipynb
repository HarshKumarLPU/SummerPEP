```python
s = {10,20,30}
s.add(40)         # it adds element in first
print(s)
s.update([50,60]) # add elements randomly
print(s)
s.remove(20)      #remove will not give error if element is not present
print(s)
s.discard(100)    #discard will not give error if element is not present 
print(s)
print(s.pop())    # pop out the first element
print(s)
copy = s.copy()
print(copy)
copy.clear()      # clear will use to clear the but will keep the structure
print(copy)
```
