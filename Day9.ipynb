```python
def countdown(n):
    if n==0:
        return
    print(n)
    countdown(n-1)

countdown(5)
```

```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5));
```

```python
def sum(n):
    if n==1:
        return 1
    return n+sum(n-1)
print(sum(5))
```

```python
def list_sum(lst):
    if len(lst) == 0:      
        return 0
    return lst[0] + list_sum(lst[1:]) 

numbers = [10, 20, 30, 40, 50]
print("Sum =", list_sum(numbers))
```

```python
def largest(lst):
    if len(lst) == 1:
        if type(lst[0]) == list:
            return largest(lst[0])
        return lst[0]

    first = largest([lst[0]]) if type(lst[0]) == list else lst[0]
    rest = largest(lst[1:])

    if first > rest:
        return first
    else:
        return rest


data = [10, [25, 5], [7, [80, 12]], 30]

print("Largest =", largest(data))
```

```python
def calculator(operation):

    def add(a, b):
        return a + b

    def sub(a, b):
        return a - b

    def mul(a, b):
        return a * b

    def div(a, b):
        return a / b

    if operation == "+":
        return add
    elif operation == "-":
        return sub
    elif operation == "*":
        return mul
    elif operation == "/":
        return div


calc = calculator("*")
print(calc(10, 5))
```

```python
def my_map(func, data):
    result = []
    for i in data:
        result.append(func(i))
    return result

# Function
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

print(my_map(square, numbers))
```

```python
def compress(s):
    if len(s) == 0:
        return ""

    count = 1
    while count < len(s) and s[count] == s[0]:
        count += 1

    return s[0] + str(count) + compress(s[count:])


text = "aaabbbcccdda"
print(compress(text))
```

```python
def number_to_words(n):
    words = ["zero", "one", "two", "three", "four",
             "five", "six", "seven", "eight", "nine"]

    if len(n) == 0:        
        return

    print(words[int(n[0])], end=" ")
    number_to_words(n[1:])


num = input("Enter a number: ")
number_to_words(num)
```

```python
def my_reduce(func, data):
    result = data[0]

    for i in data[1:]:
        result = func(result, i)

    return result


numbers = [1, 2, 3, 4, 5]

ans = my_reduce(lambda x, y: x + y, numbers)

print(ans)
```
