```python
students = {
    "Harsh": 85,
    "Rahul": 72,
    "Aman": 90,
    "Neha": 68,
    "Priya": 80
}

average = sum(students.values()) / len(students)

print("Class Average:", average)
print("Students Scoring Above Average:")

for name, marks in students.items():
    if marks > average:
        print(name)
```

```python
network = {
    "Harsh": {"Rahul", "Aman", "Neha"},
    "Rahul": {"Harsh", "Aman", "Priya"},
    "Aman": {"Harsh", "Rahul"},
    "Neha": {"Harsh"},
    "Priya": {"Rahul"}
}

print("Harsh's Friends:", network["Harsh"])

mutual = network["Harsh"] & network["Rahul"]
print("Mutual Friends of Harsh and Rahul:", mutual)

suggestion = network["Rahul"] - network["Harsh"] - {"Harsh"}
print("Friend Suggestion for Harsh:", suggestion)
```

```python
airline = {
    "Booked": {"A1", "A2", "A3", "A4", "A5", "A6"},
    "Not Booked": {"A7", "A8", "A9", "A10"}
}

seat = input("Enter Seat Number: ")

if seat in airline["Booked"]:
    print("Double Booking! Seat already booked.")
else:
    airline["Booked"].add(seat)
    airline["Not Booked"].discard(seat)
    print("Seat booked successfully")

print("Booked Seats:", airline["Booked"])

```

```python
scoreboard = {
    "Harsh": 80,
    "Rahul": 70,
    "Aman": 90
}

name = input("Enter Participant Name: ")
score = int(input("Enter New Score: "))

scoreboard[name] = score

print("Updated Scoreboard:")

for name, score in sorted(scoreboard.items(), key=lambda x: x[1], reverse=True):
    print(name, ":", score)
```

```python
assignment1 = {"python", "java", "html", "css", "sql"}
assignment2 = {"python", "java", "css", "c++", "mysql"}

common = assignment1 & assignment2
total = assignment1 | assignment2

similarity = (len(common) / len(total)) * 100

print("Common Words:", common)
print("Unique Words:", total)
print("Similarity Percentage:", similarity, "%")
```
