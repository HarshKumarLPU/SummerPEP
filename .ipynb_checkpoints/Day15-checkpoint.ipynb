```python
file = open("my_intro.txt", "r")

lines = file.readlines()

print("Number of Lines:", len(lines))

file.close()
```

```python
file = open("my_intro.txt", "r")

content = file.read()

words = content.split()

print("Number of Words:", len(words))

file.close()
```

```python
file = open("my_intro.txt", "r")

content = file.read()

word = input("Enter word to search: ")

if word in content:
    print("Word Found")
else:
    print("Word Not Found")

file.close()
```

```python
file1 = open("my_intro.txt", "r")

content = file1.read()

file2 = open("my_intro_copy.txt", "w")

file2.write(content)

file1.close()
file2.close()

print("File Copied Successfully")
```

```python
file = open("students.txt", "w")

for i in range(5):
    name = input("Enter Student Name: ")
    file.write(name + "\n")

file.close()

print("Names Stored Successfully")
```

```python
file = open("students.txt", "r")

lines = file.readlines()

for i in range(0, len(lines), 2):
    print(lines[i])

file.close()
```

```python
class Student:
    college = "LPU"

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name,self.age,self.college)

s1 = Student("Rahul",20)
s2 = Student("Amit",21)
s1.display()
s2.display()
```
