```python
cart = ["Laptop", "Keyboard", "Mouse"]

print("Cart:", cart)

cart.append("Monitor")
print("After adding Monitor:", cart)

cart.remove("Mouse")
print("After removing Mouse:", cart)

cart[1] = "Mechanical Keyboard"
print("After update:", cart)

# Total number of products (len)
print("Total products:", len(cart))
```

```python
seats = ["A1", "A2", "A4", "A5"]

print("Before Insert:", seats)
seats.insert(2, "A3")

print("After Insert:", seats)
```

```python
orders = ["Pizza", "Burger", "Pasta", "Cold Coffee"]

print("Current Orders:", orders)
cancelled_order = orders.pop()

print("Cancelled Order:", cancelled_order)
print("Remaining Orders:", orders)
```

```python
inventory = ["Mouse", "Keyboard", "Monitor", "CPU", "Laptop"]

print("Original Inventory:", inventory)
inventory.sort()
print("Alphabetical Order:", inventory)

inventory.reverse()
print("Reverse Alphabetical Order:", inventory)
```

```python
students = ["Aman", "Rahul", "Priya", "Neha"]
backup = students.copy()

print("Original List:", students)
print("Backup List:", backup)

backup.append("Harsh")

print("\nAfter Modifying Backup")
print("Original List:", students)
print("Backup List:", backup)
```

```python
students = ["Aman", "Rahul", "Priya", "Rahul", "Neha"]


position = students.index("Aman")
print("Aman is at index:", position)

occurrences = students.count("Rahul")
print("Rahul appears", occurrences, "times")
```

```python
books = ["Python", "Java", "C++", "DBMS"]

print("Original Library:", books)

books.append("Data Science")
print("After Adding Book:", books)

print("Python is at index:", books.index("Python"))

books.remove("Java")
print("After Issuing Java:", books)

books.append("Java")
print("After Returning Java:", books)

backup = books.copy()
print("Backup Library:", backup)

books.sort()
print("Sorted Library:", books)

books.reverse()
print("Reverse Alphabetical Library:", books)

print("Final Library:", books)
```
