```python
t = ()
for i in range(5):
    value = input("")
    t = t + (value,)

print("\nElements of the Tuple are: ")
for i in range(5):
    print("At index",i,":",t[i]) 
    
    
```

```python
t = ()
for i in range(5):
    value = input("")
    t = t + (value,)

print("\nElements of the Tuple are: ")
for i in reversed(t):
    print(i)
```
