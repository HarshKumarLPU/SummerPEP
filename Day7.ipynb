```python
# A bank wants to automate the calculation of simple interest for its customer. WRITE a program using a function that calculates the interset based on the required inputs 

def calculate_interest(principal, rate, time):
    interest = (principal * rate * time) / 100
    return interest


principal = float(input("Enter Principal Amount: "))
rate = float(input("Enter Rate of Interest: "))
time = float(input("Enter Time (in years): "))


si = calculate_interest(principal, rate, time)


print("Simple Interest =", si)
```

```python
# A weather application stores the temperature in celsius, but ussers want to view it in farenheit.Write a program using a function to perform the conversion

def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = celsius_to_fahrenheit(celsius)

print("Temperature in Fahrenheit =", fahrenheit)
```

```python
#A shopping mall charges GST on every purchase. Write a program that calculates the fincal bill amount after adding GST. Test your solution for multiple customers

def calculate_bill(amount, gst_percent):
    gst = (amount * gst_percent) / 100
    final_bill = amount + gst
    return gst, final_bill

customers = int(input("Enter the number of customers: "))

for i in range(1, customers + 1):
    print(f"\nCustomer {i}")

    amount = float(input("Enter Purchase Amount: ₹"))
    gst_percent = float(input("Enter GST Percentage: "))

    gst, final_bill = calculate_bill(amount, gst_percent)

    print("GST Amount: ₹", gst)
    print("Final Bill Amount: ₹", final_bill)
```

```python
# An online scholarship portal accepts applicatons only from eligible candidates. Write a function that validates whether A USER IS ELIGIBLE  based on age and returns an appropriate message 

def check_eligibility(name, age):
    if age >= 18:
        return name + " is Eligible for Scholarship Application."
    else:
        return name + " is Not Eligible for Scholarship Application."


name = input("Enter Candidate Name: ")
age = int(input("Enter Candidate Age: "))

result = check_eligibility(name, age)

print("\nResult:")
print(result)
```

```python
# Create a function that can accept any number of details about a family such as name, age, city, occupation, moblie number, etc., and display all the information povided.


def family_details(**details):
    print("\n----- Family Details -----")

    for key, value in details.items():
        print(key, ":", value)


name = input("Enter Name: ")
age = int(input("Enter Age: "))
city = input("Enter City: ")
occupation = input("Enter Occupation: ")
mobile = input("Enter Mobile Number: ")

family_details(
    Name=name,
    Age=age,
    City=city,
    Occupation=occupation,
    Mobile=mobile
)
```

```python
# A supermarket wants to generate the final bill for its customera. The total amount depends on the price, quantity purchased and an optional discount. Write a function to calculate the fincal payable amount 

def calculate_bill(price, quantity, discount=0):
    total = price * quantity
    discount_amount = (total * discount) / 100
    final_bill = total - discount_amount
    return final_bill


price = float(input("Enter Price: "))
quantity = int(input("Enter Quantity: "))

choice = input("Do you have discount? (yes/no): ")

if choice.lower() == "yes":
    discount = float(input("Enter Discount Percentage: "))
    bill = calculate_bill(price, quantity, discount)
else:
    bill = calculate_bill(price, quantity)

print("\nFinal Bill =", bill)
```

```python
# Write a program that accepts a student's name and marks in multiple subjects. Using a functions , calculate the total marks, average marks and display whether a Student has passsed or failed.


def student_result(name, marks):
    total = sum(marks)
    average = total / len(marks)

    if average >= 40:
        status = "Passed"
    else:
        status = "Failed"

    print("\nStudent Name:", name)
    print("Total Marks:", total)
    print("Average Marks:", average)
    print("Result:", status)


name = input("Enter Student Name: ")

n = int(input("Enter Number of Subjects: "))

marks = []

for i in range(n):
    mark = float(input(f"Enter Marks of Subject {i+1}: "))
    marks.append(mark)

student_result(name, marks)
```

```python
# An organization wants to calculate the salary of its employees. the fnal salary depends on the basic salary along with the HRA and DA allowances. Write a function that calculates and return the final salary.


def calculate_salary(basic_salary, hra_percent, da_percent):
    hra = (basic_salary * hra_percent) / 100
    da = (basic_salary * da_percent) / 100

    final_salary = basic_salary + hra + da
    return final_salary

basic_salary = float(input("Enter Basic Salary: "))
hra_percent = float(input("Enter HRA Percentage: "))
da_percent = float(input("Enter DA Percentage: "))

salary = calculate_salary(basic_salary, hra_percent, da_percent)

print("\nFinal Salary =", salary)
```
