```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reverse =", rev)
    
    
```

```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if num == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
num = int(input("Enter a number: "))
i = 2
prime = 1

if num <= 1:
    prime = 0
else:
    while i < num:
        if num % i == 0:
            prime = 0
            break
        i += 1

if prime == 1:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1
count = 1

while count <= n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    count += 1
```
