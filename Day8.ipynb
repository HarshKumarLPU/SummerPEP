```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Square each number
squares = list(map(lambda x: x * x, numbers))
print("Squares:", squares)

# Keep even squares
even = list(filter(lambda x: x % 2 == 0, squares))
print("Even Squares:", even)

# Sum of even squares
total = reduce(lambda x, y: x + y, even)
print("Sum:", total)
```
